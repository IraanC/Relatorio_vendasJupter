# Relatório de Vendas

```python
import pandas as pd;
import matplotlib.pyplot as plt;
import altair as alt;
```


Imports


```python
# Importação do dataFrame (df):
df = pd.read_csv('Client.csv')
# Adicão de uma nova coluna  nomeada como "Total":
df['Total'] = df['Preco_Unitario'] * df['Quantidade']
# Remove os valores invalidos ou NAN
df = df.dropna()
df
```

Total de vendas por Categoria

```python
alt.Chart(df).mark_bar().encode(x= 'Total', y='Categoria')
```

As 10 vendas com maior valor total

```python
df[[ 'Data','Nome_Produto', 'Categoria', 'Quantidade', 'Total']].head(10)
```

Maior registro de produtos de uma mesma categoria em uma venda


```python
x = df.loc[df['Categoria'] == 'Eletrônicos', ['Nome_Produto', 'Categoria', 'Quantidade']]
x['Quantidade'] = pd.to_numeric(x['Quantidade'], errors='coerce')

max_value = x['Quantidade'].max()
print(f'O maior registro de produtos da mesma categoria em única venda foi : {max_value:.0f}')
# Visualição
df.loc[df['Quantidade'] == max_value, ['Nome_Produto', 'Categoria', 'Quantidade']]
```

Média de Faturamento por venda

```python
mean = df['Total'].mean()
print(f'A Média de valor por venda {mean :.2f}')
```

Total de Faturamento

```python
total= df['Total'].sum()

card = alt.Chart(pd.DataFrame({'Texto': [f'Total : {total}']})).mark_text(
    align='center',
    baseline='middle',
    fontSize=30,
    color='black'
).encode(
    x=alt.value(0),  # Posição X
    y=alt.value(0),  # Posição Y
    text='Texto:N'
).properties(
    width=200,
    height=100
)
card.display()
```
